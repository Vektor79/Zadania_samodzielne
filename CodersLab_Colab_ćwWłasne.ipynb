Mapa CIEPLNA (heatmap):


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
numeric_tips = tips.select_dtypes(include=['number'])  # Filter for numeric columns

sns.heatmap(numeric_tips.corr(), annot=True)
plt.show()
```

Wykres PUDEŁKOWY boxplot:

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

wykres ROZKŁADU DANYCH Distribution Plot, displot:

Wizualizuje rozkład danych, często stosowany do danych ciągłych.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(tips['total_bill'], kde=True)
plt.show()
```

```python
# dataset visualization - total_bill vs. tip
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.set(style='darkgrid')
plot = sns.relplot(x='total_bill', y='tip', hue = 'sex', data=tips)
plt.show()
```

Dodanie parametru hue do metody relplot pozwoliło na uzyskanie większej liczby szczegółów o danych. Na przykład — wiemy, że wraz ze wzrostem wartości rachunku mężczyźni chętniej dają większe napiwki niż kobiety.

```python
# dataset visualization - total_bill vs. tip
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.set(style='darkgrid')
plot = sns.relplot(x='total_bill', y='tip', hue = 'smoker', data=tips)
plt.show()
```

Na powyższej wizualizacji widzimy, że z reguły osoby niepalące dają częściej wyższe napiwki w zależności od wartości rachunku. Mimo wszystko osoba paląca dała najwyższy rachunek spośród wszystkich gości znajdujących się w naszym zbiorze danych.

```python
plot.figure.savefig('tips.png')    # zapisanie/pobranie wykresu na dysk lokalny komputera
```
